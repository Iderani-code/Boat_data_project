```python
#To generate the percentage of boats present
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('boat_data.csv')

# Get the counts of each boat type
boat_type_counts = df['Boat Type'].value_counts()

# Extract labels and corresponding counts
mylabels = boat_type_counts.index
y = boat_type_counts.values

# Create the pie chart
plt.pie(y, labels=mylabels, autopct='%1.1f%%')
plt.legend(title="Boat Type:", loc="center left", bbox_to_anchor=(1, 0.5))
plt.axis('equal')
plt.show()
```

```python
#To generate the Top Ten viewed boats

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('boat_data.csv')

grouped_data = df.groupby('Boat Type')

views_counts = grouped_data['Number of views last 7 days'].max()

sorted_views = views_counts.sort_values(ascending=False)

top_10_viewed_items = sorted_views.head(10)

print(top_10_viewed_items)

myvalues = top_10_viewed_items.values

# Chart label and plotting
mylabels = top_10_viewed_items.index

plt.pie(myvalues, labels=mylabels)
plt.legend(title="Top Ten boats with highest numbers of views:" ,loc="center left", bbox_to_anchor=(1.2, 0.5))
plt.show()
```

```python
#Checking for top expensive boats

import pandas as pd

df = pd.read_csv('boat_data.csv')

df_sorted = df.sort_values(by='Price', ascending=False)

top_10_expensive_boats = df_sorted.head(10)

top_10_expensive_boats
```

```python
#Checking for top expensive boats/Graph

import pandas as pd
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='Price', ascending=False)
top_10_expensive_items = df_sorted.head(10)

plt.bar(top_10_expensive_items['Boat Type'], top_10_expensive_items['Price'], edgecolor='black')
plt.xlabel('Boat Type')
plt.ylabel('Price')
plt.title('Top 10 Most Expensive Boat')
plt.xticks(rotation=45)
plt.show()
```
